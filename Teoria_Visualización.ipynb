```python
# np.linspace(10,100,100) 
# la funcion genera valores espaciados linealmente.
# la funciÃ³n logspace genera valores espaciados logarÃ­tmicamente.
```

```python
def nlog(x):
    return np.log(x)*x

p = np.linspace(10,100,100)
plt.plot(p,nlog(p) , p,np.log(p),  p,p)

plt.show()
```

```python
p = np.linspace(10,100,100)
plt.plot(p,nlog(p),'g^' , p,np.log(p),'r--',  p,p,'bs')

plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1,t2 = np.arange(0,5,0.1),np.arange(0,5,0.02)

plt.subplot(2,1,1)
plt.plot(t1,f(t1),'bo' , t2,f(t2),'k')

plt.subplot(2,1,2)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')

plt.show()
```

```python
def grafico(d):
    x = np.array(d)
    y = x**3 - 50*x
    plt.plot(x,y)
    plt.show()

grafico(range(-7,8))
```

```python
ax = plt.subplot(111)

t = np.arange(0,5,0.01)
s = np.cos(2*np.pi*t)
line,= plt.plot(t,s, lw=2)

plt.annotate('max. local',xy=(2,1),
              xytext = (3.5,1.5),
              arrowprops= dict(facecolor='red',shrink=0.01),)
plt.ylim(-2,2)

plt.show()
```

Graficos de puntos

```python
x = np.random.rand(50)
y = x+10*np.random.rand(50)
color = np.random.randint(0,50,50)
tamano = np.abs(np.random.randn(50)) * 250


plt.scatter(x,y, c = color, s = tamano, alpha = 0.3)
plt.xlabel("coor en x")
plt.ylabel("coor en y")
plt.show()
```

Graficos de barras

```python
objetos = ('Python','R','C','Java','Scala','Perl')
x = np.arange(len(objetos))
y = [10,9,7,7,6,3]

plt.bar(x,y, align='center', alpha = 0.8)
plt.xticks(x,objetos)
plt.title("Uso de lenguajes de programacion")
plt.ylabel("Nivel de uso en la empresa")

plt.show()
```

```python
objetos = ('Python','R','C','Java','Scala','Perl')
x = np.arange(len(objetos))
y = [10,9,7,7,6,3]

plt.barh(x,y, align='center', alpha = 0.8)
plt.yticks(x,objetos)
plt.title("Uso de lenguajes de programacion")
plt.xlabel("Nivel de uso en la empresa")

plt.show()
```

```python
carne , pescado = (90,55,40,65) ,(82,62,54,20)
grupos = 4

fig , ax = plt.subplots()

i= np.arange(grupos)

anchura_barra , opacidad = 0.35, 0.3

r1 = plt.bar(i, carne, anchura_barra, alpha = opacidad, color = 'r', label = 'carne')
r2 = plt.bar(i, pescado, anchura_barra, alpha = opacidad, color = 'b', label = 'pescado')

plt.title("Preferencia por provincia")
plt.xlabel("Provincia")
plt.ylabel("Puntuacion")

plt.xticks(i, ('Madrid','Barcelona','Sevilla','Galicia'))
plt.legend()

plt.tight_layout()

plt.show()
```
