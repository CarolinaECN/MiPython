#### Numero de cluster optimo: metodo del dendrograma

```python
import scipy.cluster.hierarchy as sch # es un grafico de arbol que nos ayuda a determinar el numero optimo de clusters
```

```python
dendrograma = sch.dendrogram(sch.linkage(x,method='ward'))
    
plt.title('Metodo del Dendrograma')
plt.xlabel('Nro clientes')
plt.ylabel('Distancia Euclidiana') # distancia entre dos puntos en un plano
plt.show()
```

* En la distancia (vacio) mayor se traza una horizontal y se cuenta el numero de cortes.
* En este caso la horizontal estaria por ejemplo en 200. Son 5 cortes y por tanto 5 clusters como numero optimo. 
* Coinicidiendo con el metodo del codo. 

<img src="InterpretacionDendrograma.png">

#### Creacion del modelo de Clustering Jerarquico

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
hc
```

#### Prediccion/Segmentacion

```python
y_prediccion = hc.fit_predict(x)
y_prediccion
```

Por cada valor de X nos esta indicando a que clase pertenece. Como se ve las clases van de 0 a 4, total 5 clases.

##### Evaluacion del modelo NO aplica en Clustering

#### Visualizacion del Modelo

```python
plt.scatter(x[y_prediccion==0, 0], x[y_prediccion==0, 1], s=100, c = 'red',label='Cluster 1')
plt.scatter(x[y_prediccion==1, 0], x[y_prediccion==1, 1], s=100, c = 'blue',label='Cluster 2')
plt.scatter(x[y_prediccion==2, 0], x[y_prediccion==2, 1], s=100, c = 'green',label='Cluster 3')
plt.scatter(x[y_prediccion==3, 0], x[y_prediccion==3, 1], s=100, c = 'cyan',label='Cluster 4')
plt.scatter(x[y_prediccion==4, 0], x[y_prediccion==4, 1], s=100, c = 'magenta',label='Cluster 5')
plt.title('Clustering de clientes')
plt.xlabel('Ingresos en K$')
plt.ylabel('Puntuacion de Gasto (1-100)')
plt.legend()
plt.show()
```

Estos son los 5 grupos/segmentos que se obtienen del total de clientes.
